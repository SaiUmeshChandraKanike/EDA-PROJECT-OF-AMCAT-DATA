# FIND THE OUTLIERS EACH NUMERICAL COLUMN

OUTLIERS USING IQR

```python
def find_outliers(column):
    Q1 = amcat[column].quantile(0.25)
    Q3 = amcat[column].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = amcat[(amcat[column] < lower_bound) | (amcat[column] > upper_bound)]
    return outliers
```

OUTLIERS OF EACH NUMERICAL COLUMN

```python
outliers_dict = {}
for col in numerical_cols:
    outliers = find_outliers(col)
    outliers_dict[col] = len(outliers)
    print(f"outliers in {col} : {len(outliers)}")
```
